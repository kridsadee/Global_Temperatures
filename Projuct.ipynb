# Linear Regression 

## Define x and y 

```python
data_m.head()
```

```python
data_m.drop('Date',axis=1,inplace=True)
data_m.head()
```

```python
data_m
```

```python
x = data_m.drop(['Land/Ocean Tem Un'],axis=1).values
y = data_m['Land/Ocean Tem Un'].values
```

```python
print(x)
```

```python
print(y)
```

#### Split the dataset in training set and test set 

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

#### Train the model on the training set 

```python
from sklearn.linear_model import LinearRegression
ml=LinearRegression()
ml.fit(x_train,y_train)
```

#### Predict the test set resulte 

```python
y_pred=ml.predict(x_test)
print(y_pred)
```

```python
ml.predict([[0.749,1.105,8.242,1.738,-3.206,2.822,12.833]])
```

#### Evaluate the model 

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
plt.figure(figsize=(15,10))
plt.scatter(y_test,y_pred)
plt.xlabel('Land And Ocean Average Temperature Uncertainty')
plt.ylabel('Land Average Temperature')
```
